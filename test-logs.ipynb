```python
import numpy as np
import matplotlib.pyplot as plt

# Extract the data from the logs.txt file
time_taken = []
with open('logs.txt', 'r') as f:
    for line in f:
        try:
            # Extract the time taken from each line
            time_taken.append(int(line.split(':')[1].strip()))
        except ValueError:
            print(f"Skipping invalid line: {line.strip()}")

# Creating the boxplot on a logarithmic scale
plt.figure(figsize=(10, 6))
plt.boxplot(time_taken)
plt.yscale('log')
plt.title("Boxplot on Logarithmic Scale")
plt.ylabel("Time taken (log scale)")
plt.grid(True, which="both", linestyle="--", linewidth=0.5)
plt.show()
```
