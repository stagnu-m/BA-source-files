# Prepare Data

```python
import numpy as np
import matplotlib.pyplot as plt

# Prepare Data
Standard_L_1 = []
with open('Data/Standard_L_1.txt', 'r') as f:
    for line in f:
        try:
            # Extract the time taken from each line
            Standard_L_1.append(int(line.split(':')[1].strip()))
        except ValueError:
            print(f"Skipping invalid line: {line.strip()}")

MinMax_L_1 = []
with open('Data/MinMax_L_1.txt', 'r') as f:
    for line in f:
        try:
            # Extract the time taken from each line
            MinMax_L_1.append(int(line.split(':')[1].strip()))
        except ValueError:
            print(f"Skipping invalid line: {line.strip()}")

Standard_L_1_pro_step = []
with open('Data/Standard_L_1_pro_step.txt', 'r') as f:
    for line in f:
        try:
            # Extract the time taken from each line
            Standard_L_1_pro_step.append(int(line.split(':')[1].strip()))
        except ValueError:
            print(f"Skipping invalid line: {line.strip()}")

Standard_L_1_after_solution = []
with open('Data/Standard_L_1_after_solution.txt', 'r') as f:
    for line in f:
        try:
            # Extract the time taken from each line
            Standard_L_1_after_solution.append(int(line.split(':')[1].strip()))
        except ValueError:
            print(f"Skipping invalid line: {line.strip()}")

# # Plotting the boxplot
# plt.figure(figsize=(10, 6))
# plt.boxplot(time_taken[:1000], whis=[0, 100])  # Whiskers set to min and max
# plt.yscale('linear')  # Logarithmic scale
# plt.title("Boxplot with Min-Max Whiskers on a Logarithmic Scale")
# plt.ylabel("Time taken (log scale)")
# plt.grid(True, which="both", linestyle="--", linewidth=0.5)
# plt.show()
```

# Standard $L_1$

```python
# Calculate the average
Standard_L_1_average = np.mean(Standard_L_1[:1000])

# Plotting the histogram
plt.figure(figsize=(10, 6))
plt.hist(Standard_L_1[:1000], bins=30, color='blue', edgecolor='black', alpha=0.7)
plt.axvline(Standard_L_1_average, color='red', linestyle='--', label=f'Average: {Standard_L_1_average:.2f}ms')
plt.title("Standard $L_1$")
plt.xlabel("Time taken in milliseconds")
plt.ylabel("Frequency")
plt.legend()
plt.grid(True, linestyle="--", linewidth=0.5)
plt.savefig("Results/Standard_L_1_plot.png", dpi=300, bbox_inches="tight")  # File name and format
plt.show()
```
